Build a basic Deep Learning Model to predict data points using Make Circle Dataset     

```
#Import libraries/packages
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```
#Load Dataset
from sklearn.datasets.samples_generator import make_circles
```

```
# Load Feature and Target variables
X, y = make_circles(n_samples=10000,
                    noise=0.1,
                    factor=0.2,
                    random_state=0)
print(X)
print(y)
```

```
#Plot data
plt.figure(figsize=(5, 5))
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses")
```

```
#Import Keras models
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
```

```
# Design your DL Model
model = Sequential()
model.add(Dense(4, input_shape=(2,), activation='tanh'))
model.add(Dense(1, activation='sigmoid'))
```

```
model.summary()
```

```
#Compile yout model with Optimizer, Learning rate, Loass, Metrics 
#help(model.compile)
model.compile(SGD(lr=0.5), 'binary_crossentropy', metrics=['accuracy'])
```

```
#Fit your train data in to DL Model
model.fit(X, y, epochs=20)
```

```
#Generate Test Data
hticks = np.linspace(-1.5, 1.5, 101)
vticks = np.linspace(-1.5, 1.5, 101)
aa, bb = np.meshgrid(hticks, vticks)
ab = np.c_[aa.ravel(), bb.ravel()]
plt.plot(aa.ravel(),bb.ravel())
```

```
#Prediction
c = model.predict(ab)
cc = c.reshape(aa.shape)
```

```
#Plot the results
plt.figure(figsize=(5, 5))
plt.contourf(aa, bb, cc, cmap='bwr', alpha=0.2)
plt.plot(X[y==0, 0], X[y==0, 1], 'ob', alpha=0.5)
plt.plot(X[y==1, 0], X[y==1, 1], 'xr', alpha=0.5)
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.legend(['0', '1'])
plt.title("Blue circles and Red crosses")
```
